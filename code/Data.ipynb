### 1. Extracting midi files from dataset "lmd_matched" (which was found in web, contains well-known songs without voice) 

### Initial dataset containes 6 categories of emotions, 100 random samples from each category were selected

```python
import pandas as pd
from glob import glob
import numpy as np
import json
from tqdm import tqdm
import os
import random
from joblib import Parallel, delayed
import music21
from music21 import converter, stream, instrument, note, chord
import pickle
```

```python
files_list = glob('lmd_matched/**/*.mid', recursive=True)
```

```python
with open('labels.json') as file:
    labels = json.load(file)
```

```python
files_1 = []
files_2 = []
files_3 = []
files_4 = []
files_5 = []
files_6 = []
for file in tqdm(files_list):
    key = os.path.split(os.path.dirname(file))[-1]
    try:
        label = labels[key]
    except:
        continue
    if label == 1:
        files_1.append(file)
    if label == 2:
        files_2.append(file)
    if label == 3:
        files_3.append(file)
    if label == 4:
        files_4.append(file)
    if label == 5:
        files_5.append(file)
    if label == 6:
        files_6.append(file)
```

```python
new_files_1 = []
new_files_2 = []
new_files_3 = []
new_files_4 = []
new_files_5 = []
new_files_6 = []

lists = [new_files_1, new_files_2, new_files_3, new_files_4, new_files_5, new_files_6]
old_lists = [files_1, files_2, files_3, files_4, files_5, files_6]
for index, old_list in enumerate(old_lists):
    lists[index] = np.array(old_list)[np.random.choice(len(old_list), 100, replace=False)]
```

```python
import shutil
import os
index = 0
for l, list_ in enumerate(lists): 
    index += 1000
    for path in list_:
        shutil.copy2(path, 'data/') 
        name = path.split('/')[-1]
        print(name)
        os.rename('data/{}'.format(name), 'data/{}.mid'.format(index))
        index += 1
    index = (l + 1) * 1000
```

### 2. As we decided to dedicate this part of the project to mono-music, the first problem we have faced with is an extracting only piano sounds from the songs.

```python
def processMidi(file, idx):
    if idx > 360:
        try:
            midi = converter.parse(file)
            parts = instrument.partitionByInstrument(midi)
            if parts:
                for i in parts.parts:
                    if len(i) > 500 and i.partName == 'Piano':
                        i.write('midi', fp='piano/{}.mid'.format(idx))
                        break
        except Exception:
            print('h')
            pass
```

```python
Parallel(n_jobs=-1, verbose=1)(delayed(processMidi)(file, idx) for idx, file in tqdm(enumerate(glob("data/*.mid"))))  
```

### 3. As wee decided to use notes (ноты) and chords (аккорды) as input, we had to convert all songs into its notes transcription. For this purpose, we tried different packages and techniques, then finally focused on package 'music21', which was recently proposed by MIT students.

```python
def extractNodes(file):
    notes = []
    number = int(file.split('.')[0].split('/')[-1])
    midi = converter.parse(file)
    for element in midi.recurse():
        if isinstance(element, note.Note):
            notes.append(str(element.pitch))
        elif isinstance(element, chord.Chord):
            notes.append('.'.join(str(n) for n in element.normalOrder))
    return (number, notes)
```

```python
result = Parallel(n_jobs=-1)(delayed(extractNodes)(file) for file in tqdm(glob("piano/*.mid")))
```
